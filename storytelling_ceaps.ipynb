## STORYTELLING - Perguntas chave

### Pergunta 1: Os gastos estﾃ｣o aumentando ao longo dos anos?
A pandemia (2020) afetou? Ano de eleiﾃｧﾃ｣o (2022) teve mais gastos?

```python
gastos_por_ano = df.groupby('ANO')['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_por_ano.columns = ['Ano', 'Total Gasto']

print("Gastos por Ano:")
print("-" * 40)
for _, row in gastos_por_ano.iterrows():
    print(f"{int(row['Ano'])}: R$ {row['Total Gasto']:>15,.2f}")
```

```python
# Grﾃ｡fico de Linha - Evoluﾃｧﾃ｣o dos Gastos por Ano
fig1 = px.line(
    gastos_por_ano,
    x='Ano',
    y='Total Gasto',
    title='沒 Evoluﾃｧﾃ｣o dos Gastos dos Senadores (2019-2022)',
    markers=True
)

fig1.update_layout(
    xaxis_title='Ano',
    yaxis_title='Total Gasto (R$)',
    yaxis_tickformat=',.0f',
    template='plotly_white',
    font=dict(size=12),
    xaxis=dict(tickmode='array', tickvals=[2019, 2020, 2021, 2022])
)

fig1.update_traces(
    line=dict(width=3, color='#1f77b4'),
    marker=dict(size=10)
)

fig1.show()
```

```python
# Salvar grﾃ｡fico como HTML (interativo)
fig1.write_html('grafico1_evolucao_anual.html')
print("笨 Grﾃ｡fico salvo: grafico1_evolucao_anual.html")
```

```python
# Exportar dados para Excel
gastos_por_ano.to_excel('excel_pergunta1_gastos_ano.xlsx', index=False)
print("笨 Excel salvo: excel_pergunta1_gastos_ano.xlsx")
```

Os gastos estﾃ｣o aumentando ao longo dos anos? 

A pandemia (2020) afetou? Ano de eleiﾃｧﾃ｣o (2022) teve mais gastos? 

CONCLUSﾃグ: Houve uma queda significativa no ano de 2020 devido ao recesso pandemico e paralizaﾃｧﾃ｣o das atividades publicas porem retomadas com crescimento entre 2021 a 2022

## Pergunta 2: Com o que os senadores mais gastam?

Quais categorias de despesa consomem mais recursos pﾃｺblicos?

```python
# Agrupar gastos por tipo de despesa
gastos_por_tipo = df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_por_tipo.columns = ['Tipo de Despesa', 'Total Gasto']
gastos_por_tipo = gastos_por_tipo.sort_values('Total Gasto', ascending=False)

# Calcular percentual
gastos_por_tipo['Percentual'] = (gastos_por_tipo['Total Gasto'] / gastos_por_tipo['Total Gasto'].sum() * 100).round(1)

print("Gastos por Tipo de Despesa:")
print("-" * 70)
for _, row in gastos_por_tipo.iterrows():
    print(f"{row['Percentual']:>5.1f}% | R$ {row['Total Gasto']:>15,.2f} | {row['Tipo de Despesa'][:100]}")
```

```python
# Grﾃ｡fico de Barras Horizontal - Gastos por Tipo
gastos_por_tipo['Tipo Curto'] = gastos_por_tipo['Tipo de Despesa'].str[:35] + '...'

fig2 = px.bar(
    gastos_por_tipo,
    x='Total Gasto',
    y='Tipo Curto',
    orientation='h',
    title='汳ｰ Com o que os Senadores mais Gastam? (2019-2022)',
    text='Percentual',
    hover_name='Tipo de Despesa'
)

fig2.update_layout(
    xaxis_title='Total Gasto (R$)',
    yaxis_title='',
    xaxis_tickformat=',.0f',
    template='plotly_white',
    font=dict(size=11),
    height=500,
    margin=dict(l=250),
    yaxis={'categoryorder':'total ascending'}
)

fig2.update_traces(
    texttemplate='%{text:.1f}%',
    textposition='outside',
    marker_color='#2ecc71'
)

fig2.show()
```

```python
# Versﾃ｣o Pizza (alternativa
fig2_pizza = px.pie(
    gastos_por_tipo,
    values='Total Gasto',
    names='Tipo Curto',  # 竊 usa nome curto
    title='汳ｰ Distribuiﾃｧﾃ｣o dos Gastos por Categoria (2019-2022)',
    hover_name='Tipo de Despesa',  # 竊 nome completo no hover
    hole=0.4
)

fig2_pizza.update_traces(
    textposition='outside',
    textinfo='percent'  # 竊 mostra sﾃｳ percentual (sem o nome)
)

fig2_pizza.update_layout(
    font=dict(size=10),
    showlegend=True,  # 竊 legenda separada
    legend=dict(
        font=dict(size=9),
        orientation='h',  # 竊 legenda horizontal
        yanchor='bottom',
        y=-0.3  # 竊 posiﾃｧﾃ｣o abaixo do grﾃ｡fico
    ),
    margin=dict(b=150)  # 竊 espaﾃｧo embaixo para legenda
)

fig2_pizza.show()
```

```python
# Salvar grﾃ｡ficos
fig2.write_html('grafico2_gastos_tipo.html')
fig2_pizza.write_html('grafico2_pizza_tipo.html')
print("笨 Grﾃ｡ficos salvos!")
```

```python
# Exportar para Excel
gastos_por_tipo.to_excel('excel_pergunta2_gastos_tipo.xlsx', index=False)
print("笨 Excel salvo: excel_pergunta2_gastos_tipo.xlsx")
```

Com o que os senadores mais gastam?

Quais categorias de despesa consomem mais recursos pﾃｺblicos?

CONCLUSﾃグ: Consultorias, assesorias ou seja cabide de empregos seguido por despesas de viagens, hospedagens e alimentaﾃｧﾃ｣o

## Pergunta 3: Quais os 10 senadores que mais gastaram?

Identificar os maiores utilizadores da cota parlamentar.

```python
# Agrupar gastos por senador
gastos_por_senador = df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_por_senador.columns = ['Senador', 'Total Gasto']
gastos_por_senador = gastos_por_senador.sort_values('Total Gasto', ascending=False)

# Top 10
top10_senadores = gastos_por_senador.head(10)

print("沛 Top 10 Senadores que mais Gastaram (2019-2022):")
print("-" * 50)
for i, row in enumerate(top10_senadores.itertuples(), 1):
    print(f"{i:>2}. {row.Senador:<30} R$ {row._2:>15,.2f}")
```

```python
# Grﾃ｡fico de Barras Horizontal - Top 10 Senadores
fig3 = px.bar(
    top10_senadores,
    x='Total Gasto',
    y='Senador',
    orientation='h',
    title='沛 Top 10 Senadores que mais Gastaram (2019-2022)',
    text='Total Gasto'
)

fig3.update_layout(
    xaxis_title='Total Gasto (R$)',
    yaxis_title='',
    xaxis_tickformat=',.0f',
    template='plotly_white',
    font=dict(size=11),
    height=500,
    margin=dict(l=200),
    yaxis={'categoryorder':'total ascending'},
    xaxis=dict(range=[0, top10_senadores['Total Gasto'].max() * 1.3])
)

fig3.update_traces(
    texttemplate='R$ %{text:,.0f}',
    textposition='outside',
    marker_color='#e74c3c'
)

fig3.show()
```

```python
# Salvar grﾃ｡fico
fig3.write_html('grafico3_top10_senadores.html')
print("笨 Grﾃ｡fico salvo: grafico3_top10_senadores.html")
```

```python
# Exportar para Excel
gastos_por_senador.to_excel('excel_pergunta3_gastos_senador.xlsx', index=False)
print("笨 Excel salvo: excel_pergunta3_gastos_senador.xlsx")
```

 Quais os 10 senadores que mais gastaram?

Identificar os maiores utilizadores da cota parlamentar.

CONCLUSﾃグ: O top 10 senadores estﾃ｡ bem acirrado com todos gastando muitas fortunas no decorrer dos anos. Telmario Mota, Rogerio Carvallho e Mecias de Jesus foram os que mais gastaram.


## Pergunta 4: Quais fornecedores mais receberam dinheiro pﾃｺblico?

Identificar as empresas que mais se beneficiaram da cota parlamentar.

```python
# Agrupar por fornecedor
gastos_por_fornecedor = df.groupby('FORNECEDOR')['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_por_fornecedor.columns = ['Fornecedor', 'Total Recebido']
gastos_por_fornecedor = gastos_por_fornecedor.sort_values('Total Recebido', ascending=False)

# Top 10
top10_fornecedores = gastos_por_fornecedor.head(10)

print("沛｢ Top 10 Fornecedores que mais Receberam (2019-2022):")
print("-" * 60)
for i, row in enumerate(top10_fornecedores.itertuples(), 1):
    print(f"{i:>2}. R$ {row._2:>15,.2f} | {row.Fornecedor[:45]}")
```

```python
# Grﾃ｡fico de Barras Horizontal - Top 10 Fornecedores
fig4 = px.bar(
    top10_fornecedores,
    x='Total Recebido',
    y='Fornecedor',
    orientation='h',
    title='沛｢ Top 10 Fornecedores que mais Receberam (2019-2022)',
    text='Total Recebido'
)

fig4.update_layout(
    xaxis_title='Total Recebido (R$)',
    yaxis_title='',
    xaxis_tickformat=',.0f',
    template='plotly_white',
    font=dict(size=11),
    height=500,
    margin=dict(l=200, r=150),
    yaxis={'categoryorder':'total ascending'},
    xaxis=dict(range=[0, top10_fornecedores['Total Recebido'].max() * 1.3])
)

fig4.update_traces(
    texttemplate='R$ %{text:,.0f}',
    textposition='outside',
    marker_color='#9b59b6'
)

fig4.show()
```

```python
# Quantos senadores cada fornecedor do top 10 atende?
print("\n沒 Quantidade de senadores que cada fornecedor atende:")
print("-" * 60)
for fornecedor in top10_fornecedores['Fornecedor']:
    qtd_senadores = df[df['FORNECEDOR'] == fornecedor]['SENADOR'].nunique()
    print(f"{qtd_senadores:>3} senadores | {fornecedor[:50]}")
```

```python
# Salvar grﾃ｡fico
fig4.write_html('grafico4_top10_fornecedores.html')
print("笨 Grﾃ｡fico salvo: grafico4_top10_fornecedores.html")
```

```python
# Exportar para Excel
gastos_por_fornecedor.head(50).to_excel('excel_pergunta4_top50_fornecedores.xlsx', index=False)
print("笨 Excel salvo: excel_pergunta4_top50_fornecedores.xlsx")
```

## Pergunta 5: Existe algum padrﾃ｣o de gasto por mﾃｪs?

Identificar sazonalidade nos gastos. Meses com mais ou menos gastos?

```python
# Agrupar por mﾃｪs
gastos_por_mes = df.groupby('MES')['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_por_mes.columns = ['Mﾃｪs', 'Total Gasto']

# Adicionar nome do mﾃｪs
meses_nome = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Marﾃｧo', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}
gastos_por_mes['Nome do Mﾃｪs'] = gastos_por_mes['Mﾃｪs'].map(meses_nome)

print("Gastos por Mﾃｪs (soma de todos os anos):")
print("-" * 45)
for _, row in gastos_por_mes.iterrows():
    print(f"{row['Nome do Mﾃｪs']:<12} R$ {row['Total Gasto']:>15,.2f}")
```

```python
# Grﾃ｡fico de Barras - Gastos por Mﾃｪs
fig5 = px.bar(
    gastos_por_mes,
    x='Nome do Mﾃｪs',
    y='Total Gasto',
    title='沒 Padrﾃ｣o de Gastos por Mﾃｪs (2019-2022)',
    text='Total Gasto'
)

fig5.update_layout(
    xaxis_title='Mﾃｪs',
    yaxis_title='Total Gasto (R$)',
    yaxis_tickformat=',.0f',
    template='plotly_white',
    font=dict(size=12)
)

fig5.update_traces(
    texttemplate='R$ %{text:,.0f}',
    textposition='outside',
    marker_color='#f39c12'
)

fig5.show()
```

```python
# Grﾃ｡fico extra: Evoluﾃｧﾃ｣o mensal por ano (heatmap)
gastos_mes_ano = df.groupby(['ANO', 'MES'])['VALOR_REEMBOLSADO'].sum().reset_index()
gastos_mes_ano_pivot = gastos_mes_ano.pivot(index='MES', columns='ANO', values='VALOR_REEMBOLSADO')

fig5_heatmap = px.imshow(
    gastos_mes_ano_pivot,
    labels=dict(x='Ano', y='Mﾃｪs', color='Gasto (R$)'),
    title='泓難ｸ Mapa de Calor: Gastos por Mﾃｪs e Ano',
    aspect='auto',
    color_continuous_scale='Reds'
)

fig5_heatmap.update_layout(
    yaxis=dict(tickmode='array', tickvals=list(range(1,13)), ticktext=list(meses_nome.values())),
    xaxis=dict(
        tickmode='array',
        tickvals=[2019, 2020, 2021, 2022],
        ticktext=['2019', '2020', '2021', '2022']
    )
)

fig5_heatmap.show()
```

```python
# Salvar grﾃ｡ficos
fig5.write_html('grafico5_gastos_mes.html')
fig5_heatmap.write_html('grafico5_heatmap_mes_ano.html')
print("笨 Grﾃ｡ficos salvos!")
```

```python
# Exportar para Excel
gastos_por_mes.to_excel('excel_pergunta5_gastos_mes.xlsx', index=False)
gastos_mes_ano_pivot.to_excel('excel_pergunta5_gastos_mes_ano.xlsx')
print("笨 Excel salvo!")
```

Existe algum padrﾃ｣o de gasto por mﾃｪs?

Identificar sazonalidade nos gastos. Meses com mais ou menos gastos?

CONCLUSﾃグ: Percebe-se uma crescente de gastos no mes de dezembro de todos os anos e no ano de 2022 retomando as atividades pos pandemia houve um aumento expressivo dos gastos em todos os meses
